```python


class queryAtlas:
    '''
    atlasPaths : List of paths to the nii files to be considered in order of preference
    atlasLabelsPaths : List of paths to the xml files containing labels to be considered in order of preference 
    coord : List [x,y,z] of MNI coordinates
    '''
    def __init__(self,atlasPaths,atlasLabelsPaths):
        self.atlasPaths = atlasPaths
        self.atlasLabelsPaths = atlasLabelsPaths
        
    
    def MNI2XYZ1mm(self, mni):
        x =  - mni[0] + 90
        y = mni[1] + 126
        z = mni[2] + 72
        return [x,y,z]
    
    def MNI2XYZ2mm(self, mni):   
        x =  math.floor((- mni[0] + 90)/2.0)
        y = math.floor((mni[1] + 126)/2.0)
        z = math.floor((mni[2] + 72)/2.0)
        return [x,y,z]
    
    def roiName(self, atlasLabelsPath, roiNum):
        atlasDict = {}
        root = ET.parse(atlasLabelsPath).getroot()
        elem = root.find('data')
        for regionRow in elem.getchildren():
            roiNumber  = int(regionRow.items()[2][1]) + 1
            roiName = regionRow.text
            atlasDict[roiNumber] = roiName
#             print(roiNumber,roiName)
        return atlasDict[roiNum]
     
    def getAtlasRegions(self, coordMni):
        roiNumber = None
        roiName = None
        
        atlasIndex = 0
        for index,atlasPath in enumerate(self.atlasPaths):
            if roiNumber == None:
                _atlas = nib.load(atlasPath)
                atlas = _atlas.get_data()
                if _atlas.header['pixdim'][1] == 2:
                    x,y,z = self.MNI2XYZ2mm(coordMni)
                elif _atlas.header['pixdim'][1] == 1:
                    x,y,z = self.MNI2XYZ1mm(coordMni)
                else:
                    raise Exception('Unknown Pixel Dimension', _atlas.header['pixdim'][1] )
#                 print(x,y,z)
                
                roiNumber = atlas[x,y,z]
                if roiNumber != 0:
#                     print(index,self.atlasLabelsPaths)
                    roiName = self.roiName(self.atlasLabelsPaths[index], roiNumber)
                else:
                    roiNumber = None
                    atlasIndex = atlasIndex + 1
#         assert (roiNumber != None)
        if roiNumber == None:
            roiNumber, roiName, atlasIndex = 0, None, None
            
        return int(roiNumber), roiName, atlasIndex
```

```python
class connectivityDensityCheck:
    def __init__(self,seedAtlasObj, voxelAtlasObj, logBrainPath):
        self.seedAtlasObj = seedAtlasObj
        self.voxelAtlasObj = voxelAtlasObj
        self.logBrainPath = logBrainPath
        self.logBrain = nib.load(logBrainPath).get_data()
        

    def checkDensity(self, seedMNI, voxelMNI, alpha = 1.5):
        # Get the Brainnetome atlas ROI number form seedMNI
        roi  = int(self.seedAtlasObj.getAtlasRegions(seedMNI)[0])

        # Get the HO atlas ROI number from voxelMNI [if a voxel is in both HO cortical and subcortical then give priority 
        # to cortical]

        regionIndex,_,atlasIndex = self.voxelAtlasObj.getAtlasRegions(voxelMNI)

        atlas = nib.load(self.voxelAtlasObj.atlasPaths[atlasIndex]).get_data()

        # Find the density of hyper / hypo connected voxels in the atlas

        temp = np.where(np.logical_and((atlas == regionIndex), self.logBrain[:,:,:,roi-1] >= alpha ))[0]
        numberSignificantVoxelsOC = len(temp)

        temp = np.where(np.logical_and((atlas == regionIndex), self.logBrain[:,:,:,roi-1] <= -1 * alpha ))[0]
        numberSignificantVoxelsUC = len(temp)


        totalNumberVoxels = len(np.where(atlas == regionIndex)[0])

        # TEST ---------------------------------------------
        
        return numberSignificantVoxelsUC, numberSignificantVoxelsOC, totalNumberVoxels
    
    def densityCheckCSV(self, csvPath, alpha=1.5):
            # read CSV file
        df = pd.read_csv(csvPath)
        dfMatrix = df.as_matrix(['SeedName', 'SeedMNI', 'UnderConnectivityName', 'UnderConnectivityMNI', 'OverConnectivityName',\
             'OverConnectivityMNI'])

        df_new = pd.DataFrame(columns=['SeedName','SeedNameHO', 'SeedMNI', 'UnderConnectivityName',\
'UnderConnectivityNameHO', 'UnderConnectivityMNI','ConsistencyUCVoxels', 'ConsistencyUCVoxelsDensity','totalVoxelsUC',  \
             'OverConnectivityName', 'OverConnectivityNameHO', 'OverConnectivityMNI','ConsistencyOCVoxels',\
             'ConsistencyOCVoxelsDensity', 'totalVoxelsOC'])
        
        for row in tqdm(dfMatrix):
            seedName, seedMNI, UnderConnectivityName, UnderConnectedMNI,OverConnectivityName, OverConnectedMNI = row
#             print(seedName, seedMNI, UnderConnectivityName, UnderConnectedMNI,OverConnectivityName, OverConnectedMNI)
            
            UnderConnectivityNameHO = None
            ConsistencyUCVoxels = None
            ConsistencyUCVoxelsDensity = None
            OverConnectivityNameHO = None
            ConsistencyOCVoxels = None
            ConsistencyOCVoxelsDensity = None
            totalVoxels = None

            
            
            seedMNI = seedMNI.strip('\n')


            seedMNIint = seedMNI.split(' ')
            seedMNIint = [x for x in seedMNIint if x != '']

#             print('Seed: ',seedMNIint)

            seedMNIint = list(map(int, seedMNIint))
            
            seedNameHO = self.voxelAtlasObj.getAtlasRegions(seedMNIint)
#             print('seedNameHO is ',seedNameHO)

            if not pd.isna(UnderConnectedMNI):
                UnderConnectedMNI = UnderConnectedMNI.strip('\n')
                UnderConnectedMNIint = UnderConnectedMNI.split(' ')
                UnderConnectedMNIint = [x for x in UnderConnectedMNIint if x != '']
#                 print(UnderConnectedMNIint)
                UnderConnectedMNIint = list(map(int, UnderConnectedMNIint))
                
                _,UnderConnectivityNameHO,_ = self.voxelAtlasObj.getAtlasRegions(UnderConnectedMNIint)
                 
                
                ConsistencyUCVoxels, _, totalVoxelsUC =\
                self.checkDensity(seedMNIint, UnderConnectedMNIint, alpha)
                
                ConsistencyUCVoxelsDensity = ConsistencyUCVoxels/totalVoxelsUC
                

            else:
                ConsistencyUCVoxels,ConsistencyUCVoxelsDensity, totalVoxelsUC =\
                UnderConnectedMNI,ConsistencyUCVoxelsDensity, UnderConnectedMNI

            if not pd.isna(OverConnectedMNI):
                OverConnectedMNI = OverConnectedMNI.strip('\n')
                OverConnectedMNIint = OverConnectedMNI.split(' ')
                OverConnectedMNIint = [x for x in OverConnectedMNIint if x != '']
#                 print(OverConnectedMNIint)
                OverConnectedMNIint = list(map(int, OverConnectedMNIint))
                
                _,OverConnectivityNameHO,_  = self.voxelAtlasObj.getAtlasRegions(OverConnectedMNIint)

                _,ConsistencyOCVoxels, totalVoxelsOC = self.checkDensity(seedMNIint, OverConnectedMNIint, alpha)
                ConsistencyOCVoxelsDensity = ConsistencyOCVoxels/totalVoxelsUC

            else:
                ConsistencyOCVoxels, ConsistencyOCVoxelsDensity, totalVoxelsOC =\
                OverConnectedMNI, OverConnectedMNI, OverConnectedMNI

            


            df_new = df_new.append({'SeedName':seedName,
                                    'SeedNameHO': seedNameHO,
                                    'SeedMNI':seedMNI,
                                    'UnderConnectivityName':UnderConnectivityName,
                                    'UnderConnectivityNameHO':UnderConnectivityNameHO,
                                    'UnderConnectivityMNI':UnderConnectedMNI,
                                    'ConsistencyUCVoxels' : ConsistencyUCVoxels,
                                    'ConsistencyUCVoxelsDensity': ConsistencyUCVoxelsDensity,
                                    'totalVoxelsUC': totalVoxelsUC,
                                    'OverConnectivityName':OverConnectivityName,
                                    'OverConnectivityNameHO': OverConnectivityNameHO,
                                    'OverConnectivityMNI':OverConnectedMNI,
                                    'ConsistencyOCVoxels' : ConsistencyOCVoxels,
                                    'ConsistencyOCVoxelsDensity':ConsistencyOCVoxelsDensity,
                                    'totalVoxelsOC' :totalVoxelsOC
              }, ignore_index=True)

        # Extract the seedMNI and UnderConnectedMNI column
        # Call self.connectivityCheck to get the code and connectivity string
        # Create another CSV file with rows - 
        # | SeedName, SeedMNI, UnderConnectedName, UnderConnectedMNI, ConsistentCheck,
        #  OverConnectedName, OverConnectedMNI, ConsistentCheck  |

        fileName = 'connectivityDensityResults.csv'
        df_new.to_csv(fileName)

        return df_new, fileName  


    
```

```python
import pandas as pd

class queryBrainnetomeROI:
    def __init__(self, atlas_path, atlasRegionsDescrpPath):
        self.atlas_path = atlas_path
        self.numSeeds = None
        self.atlasRegionsDescrpPath = atlasRegionsDescrpPath
#         self.ROI_dictionary = None

    def MNI2XYZ1mm(self, mni):
        x =  - mni[0] + 90
        y = mni[1] + 126
        z = mni[2] + 72
        return [x,y,z]
    
    def MNI2XYZ2mm(self, mni):   
        x =  math.floor((- mni[0] + 90)/2.0)
        y = math.floor((mni[1] + 126)/2.0)
        z = math.floor((mni[2] + 72)/2.0)
        return [x,y,z]
    
    def brainnetomeQuery(self):

        atlas = nib.load(self.atlas_path).get_data()
        df = pd.read_excel(self.atlasRegionsDescrpPath)

        self.numSeeds = int(np.max(atlas))

        df = df.as_matrix(['Lobe','Gyrus','Label ID.L', 'Label ID.R','Left and Right Hemisphere','Unnamed: 5', 'lh.MNI(X,Y,Z)', 'rh.MNI(X,Y,Z)'])
        seedInfo = np.empty((self.numSeeds + 1, 5),dtype=object)

        for lobe, gyrus, i,j,region,label,lcoord,rcoord in df:
            seedInfo[i,:] = i,lobe, gyrus, region.split('_')[0] + '_' + label, lcoord
            seedInfo[j,:] = j,lobe, gyrus, region.split('_')[0] + '_' + label, rcoord

    #     seedInfo[1:]

        # Create a dictionary of seedInfo so that it can be accessed by roi number

        ROI_dictionary = {}
        for i in seedInfo[1:]:
            ROI_dictionary[i[0]] = i[1:]
            
        return ROI_dictionary


    #     d[1]

    def queryDict(self, lis): # lis: List that contains ROI numbers (1 to 246)
        
        ROI_dictionary = self.brainnetomeQuery()
        return_regions = []
        
        
        if isinstance(lis,list):
            for i in lis:
        #         return_regions.append(str(i)+"_"+d[i][0])
                try: return_regions.append(ROI_dictionary[i])
                except KeyError:
                    return_regions.append([None,None,None,None])
                    
        else:
            i = lis
            try: return_regions.append(ROI_dictionary[i])
            except KeyError:
                return_regions.append([None,None,None,None])
                
        
        return return_regions
    
    def getAtlasRegions(self, coordMni):

        _atlas = nib.load(self.atlas_path)
        atlas = _atlas.get_data()
        if _atlas.header['pixdim'][1] == 2:
            x,y,z = self.MNI2XYZ2mm(coordMni)
        elif _atlas.header['pixdim'][1] == 1:
            x,y,z = self.MNI2XYZ1mm(coordMni)
        else:
            raise Exception('Unknown Pixel Dimension', _atlas.header['pixdim'][1] )
#         print(x,y,z)

        roiNumber = atlas[x,y,z]

        lobe, gyrus, roiName, _ = self.queryDict(roiNumber)[0]
#                 if roiNumber != 0:
#                     print(index,self.atlasLabelsPaths)
#                     roiName = self.roiName(self.atlasLabelsPaths[index], roiNumber)
#                 else:
#                     roiNumber = None
#                     atlasIndex = atlasIndex + 1
#         assert (roiNumber != None)    
        return int(roiNumber),lobe, gyrus, roiName
        
```
